```python
# Dataset : Fruit Dataset=
import pandas as pd
import numpy as np
Fruits = {
    "Fruit" : ["Apple","Banana","Mango","Orange","Grapes"],
    "Price" : [1500,600,4000,2000,700],
    "Calories" :[13000,50000,20070,6000,40020],
    "Weight" : [10,50,20,60,80]
}

df = pd.DataFrame(Fruits)
print("Original Dataset : ")
print(df)

# Min-Max Normalization
# Min-Max Normalization For Price
df["Price_MinMax"] = (df["Price"] - df["Price"].min()) / (df["Price"].max() - df["Price"].min())
print("\nMin-Max Normalization : ")
print(df)

# Normalize Multiple Columns
cols = ["Calories","Weight"]
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalize Multiple Columns : ")
print(df_norm)

# Z-Score Normalization
df["Price_zScore"] = (df["Price"] - df["Price"].mean()) / df["Price"].std()
print("\nZ-Score Normalization : ")
print(df)

# Decimal Scaling
df["Price_DecimalScaling"] = df["Price"] / 10**len(str(df["Price"].max())) # NO NEED TO CHANGE 10,100,1000,... Manually. Answer will be always in zero point something.
print("\nDecimal Scaling : ")
print(df)
```

```python
# Dataset Implementation
import pandas as pd
import numpy as np

df = pd.read_csv("/content/amazon_products_dataset_Expt-14.csv")
print(df)

# Min-Max Normalization
cols = ["Price","Rating","Reviews","Units_Sold"]
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nMin-Max Normalization : ")
print(df_norm)

# Z_Score Normalization
df_zScore = (df[cols] - df[cols].mean()) / df[cols].std()
print(df_zScore)

# Decimal Scaling
df["Price_DecimalScaling"] = df["Price"] / 10**len(str((df["Price"].max()))) # NO NEED TO CHANGE 10,100,1000,10000,..... Manually. Answer will be always in zero point something.
print("\nDecimal Scaling : ")
print(df)
```

```python
# Turning Categorial Variables Into Qualitaative Variables in Python

# Lable Encoding
import numpy as np
import pandas as pd

data = {
    'Order_ID': [101, 102, 103, 104, 105, 106],
    'Customer_Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male'],
    'Payment_Method': ['UPI', 'Credit Card', 'Debit Card', 'COD', 'UPI', 'Debit Card'],
    'Product_Category': ['Electronics', 'Clothing', 'Home', 'Electronics', 'Beauty', 'Electronics'],
    'City': ['Pune', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Hyderabad'],
    'Order_Value': [4500, 2300, 3200, 5200, 1800, 4100]
}

df = pd.DataFrame(data)
print(df)

# Method 1:: Label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df['Customer_Gender_Encoded'] = le.fit_transform(df['Customer_Gender'])
print(df)

# Method 2 : One-Hot Encoding
print("\nOne-hot encoding : ")
df_encoded = pd.get_dummies(df, columns=['Payment_Method',"City"])
print(df_encoded)

# Encoding Multiple Columns
print("\nEncoding Multiple Columns : ")
df_multi = pd.get_dummies(df,columns=['Product_Category','City'])
print(df_multi)

# Method 4 : Dummy Encoding
# Default first column will removed.

df_dummy =pd.get_dummies(df,columns=['Product_Category'],drop_first=True)
print("\nDummy Encoding : ")
print(df_dummy)
```
